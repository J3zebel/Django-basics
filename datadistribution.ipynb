Normal Distribution


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from numpy import random
x = random.normal(size = (2,3))
print(x)
```

```python
x = random.normal(loc = 1, scale = 2, size = (2,3))
print(x)
```

```python
sns.displot(random.normal(loc = 1, scale = 2, size = 1000),kde=True)
plt.show()
```

Binomial Distribution

```python
x = random.binomial(n = 10, p = 0.5, size = 15)
print(x)
```

```python
sns.distplot(random.binomial(n = 10, p = 0.5, size = 1000), hist = True ,kde=True)
plt.show()
```

Comparison of normal and binomial

```python
sns.distplot(random.normal(loc = 50, scale = 5, size = 1000),hist = False,label = 'normal')
sns.distplot(random.binomial(n = 100, p = 0.5, size = 1000), hist = False ,label = 'binomial')
plt.legend()
plt.show()
```

Multinomial Distribution

```python
x = random.multinomial(n=6,pvals=[1/6,1/6,1/6,1/6,1/6,1/6],size=10)
print(x)
```

```python
sns.displot(random.multinomial(n=6,pvals=[1/6,1/6,1/6,1/6,1/6,1/6],size=10),kde=True)
plt.show()
```
