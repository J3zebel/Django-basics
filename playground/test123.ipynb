```python

df = pd.read_csv('D:\Py Django\student_marks.csv')

df["Total_Mark"] = df.select_dtypes(include=int).sum(axis=1)
df.rename(columns = {"Unnamed: 0":"Name"},inplace=True)
df
```

```python
x = df["Name"]
y = df["Total_Mark"]
plt.bar(x,y)
plt.show()
```

```python
def marks(x):
    if x >= 500:
        return x 
    
df["Total_Mark"] = df["Total_Mark"].apply(marks)
df
```
