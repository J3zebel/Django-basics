```python
df = pd.read_csv("D:\Py Django\sample.csv")

# Remove rows with any missing (NaN) values, but this doesn't modify the original DataFrame
df.dropna()

# Remove rows with any missing (NaN) values and modify the original DataFrame in place
df.dropna(inplace=True)

# Remove rows where there are missing values in the "Name" or "Model" columns
# This doesn't modify the original DataFrame, only returns a new one
df.dropna(subset=["Name","Model"])

# Replace all missing (NaN) values in the DataFrame with 0
# This doesn't modify the original DataFrame, only returns a new one
df.fillna(0)

# Replace all missing (NaN) values in the DataFrame with 0 and modify the original DataFrame
# The DataFrame 'df' is updated with NaN values replaced by 0
df.fillna(0, inplace=True)

# Replace missing (NaN) values in the "Name" column with the string "Tata"
# Modifies the "Name" column in the original DataFrame
df["Name"].fillna("Tata", inplace=True)

# Display the entire DataFrame
# Shows all rows and columns (subject to display settings)
df
```
