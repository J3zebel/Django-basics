```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score
```

"C:\\Users\91702\Downloads\employee_attrition_dataset_900.csv"

```python
df = pd.read_csv("C:\\Users\91702\Downloads\employee_attrition_dataset_900.csv")
df
```

```python
label_encoders = {}
for column in df.columns[:-1]:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])
df
```

```python
x = df.drop('attrition',axis=1)
y = df['attrition']
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=41)
x_test
```

```python
model = DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=41)
model.fit(x_train,y_train)
```

```python
predict = model.predict(x_test)
predict
```

```python
accuracy = accuracy_score(predict,y_test)
print(f"Accuracy Score : {accuracy * 100:.2f}%")
```

```python
plt.Figure(figsize=(10,13))
plot_tree(model,feature_names=['age','gender','department','education','experience_years','salary','marital_status','overtime','work_life_balance'],filled=True)
plt.show()
```
