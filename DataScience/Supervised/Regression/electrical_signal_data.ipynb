```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
df = pd.read_csv('electrical_signal_data.csv')
df
```

```python
X = df['Time (seconds)'].values
y = df['Voltage (mV)'].values
```

```python
def nl_exp_model(x,a,b,c):
    return a*np.exp(b*x)+c
```

```python
params,_ = curve_fit(nl_exp_model,X,y,p0=(3500,-0.005,100))
```

```python
y_pred = nl_exp_model(X,*params)
y_pred
```

```python
plt.scatter(X,y,label = 'Data')
plt.plot(X,y_pred,color='r')
plt.show()
```
