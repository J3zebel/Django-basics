```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("vehicle_condition.csv")
df
```

```python
x = df.drop("Condition",axis = 1)
y = df["Condition"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=44)
```

```python
model = GaussianNB()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
new_data = np.array([[98.0, 300.0, 2500.0, 2.0]])  # new sample: Temp, Pressure, RPM, Vibration
prediction = model.predict(new_data)

if prediction[0] == "Faulty":
    print(" Machine Condition: FAULTY")
else:
    print(" Machine Condition: GOOD")



plt.figure(figsize=(8,6))
plt.scatter(x_test["Oil_Pressure"], x_test["Vibration_Level"], c=(y_test == "Faulty"), cmap='coolwarm', label="Test Data")
plt.scatter(new_data[0,1], new_data[0,3], color='black', marker='*', s=200, label='New Data')
plt.xlabel("Oil Pressure")
plt.ylabel("Vibration Level")
plt.title("Machine Condition Visualization")
plt.legend()
plt.grid()
plt.show()
```
