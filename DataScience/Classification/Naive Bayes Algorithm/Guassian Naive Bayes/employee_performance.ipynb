```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
```

```python
df = pd.read_csv("D:/Py Django/DataScience/Classification/Naive Bayes Algorithm/Guassian Naive Bayes/employee_performance.csv")
df
```

```python
x = df.drop("Performance",axis = 1)
y = df["Performance"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=23)
```

```python
model = GaussianNB()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
new_data = np.array([[38.0, 14.5, 42.0, 91.0]])
prediction = model.predict(new_data)
if prediction[0] == "Poor":
    print("Predicted Performance: Poor")
else:
    print("Predicted Performance: Good")

con_matrix = confusion_matrix(y_test, y_pred)
display = ConfusionMatrixDisplay(confusion_matrix=con_matrix, display_labels=["Poor", "Good"])
display.plot(cmap='Blues')
plt.title("Confusion Matrix - Employee Performance")
plt.show()
```
