```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("patient_health_risk.csv")
df
```

```python
x = df.drop("Health_Risk",axis = 1)
y = df["Health_Risk"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=44)
```

```python
model = GaussianNB()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
new_data = np.array([[55.0, 145.0, 200.0, 65.0]])
prediction = model.predict(new_data)

if prediction[0] == "High":
    print("Predicted Health Risk: HIGH ")
else:
    print("Predicted Health Risk: LOW ")


plt.figure(figsize=(8,6))
plt.scatter(x_test["Blood_Pressure"],   x_test["Cholesterol"],   c=(y_test == "High"),cmap="coolwarm",label='Existing Data')
plt.scatter(new_data[0,1], new_data[0,2], color='black', marker='*', s=200, label='New Data')
plt.xlabel('Blood Pressure')
plt.ylabel('Cholesterol')
plt.title('Health Risk Visualization')
plt.legend()
plt.grid()
plt.show()
```
