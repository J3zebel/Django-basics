```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("supermarket_sales - Sheet1.csv")
df.drop(["Invoice ID","Branch","City","Customer type","Gender","Product line","Payment","Date","Time"],axis=1,inplace=True)
df
```

```python
x = df
x
```

```python
scalar = StandardScaler()
scalar_data = scalar.fit_transform(x)
scalar_data
```

```python
model = KMeans(n_clusters=3,random_state=22)
model.fit(scalar_data)
```

```python
label = model.labels_
label
```

```python
plt.figure(figsize=(10,6))
for i in np.unique(label):
    plt.scatter(scalar_data[label==i,0],scalar_data[label==i,6],label=f'Cluster {i}')
plt.legend()
plt.xlabel("Unit price")
plt.ylabel("Gross Income")
plt.title("Super Market Sales")
plt.show()
```
