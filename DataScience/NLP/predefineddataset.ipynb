```python
from gensim.models import Word2Vec
```

```python
import nltk
nltk.download('brown')
```

```python
from nltk.corpus import brown

sentences = brown.sents()
sentences
```

```python
model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=5,
    sg=0
)
```

```python
model.wv.most_similar("government")
```

```python
model.wv.most_similar("computer")
```
