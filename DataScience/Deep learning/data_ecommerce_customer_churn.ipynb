```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
```

```python
df = pd.read_csv("data_ecommerce_customer_churn.csv")
df.isnull().sum()
```

```python
df = df.fillna(df.mean(numeric_only=True))
df
```

```python
label_cols = ["PreferedOrderCat","MaritalStatus"]

le = LabelEncoder()
for cols in label_cols:
    df[cols] = le.fit_transform(df[cols])
```

```python
x = df.drop("Churn", axis=1).values
y = df["Churn"].values
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential([Dense(32, activation = 'relu'),
                    Dense(16, activation = 'relu'),
                    Dense(1, activation = 'sigmoid')
                    ])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
early_stop = EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)

history = model.fit(x_train,y_train,epochs=50,batch_size=32,validation_split=0.2,verbose=1,callbacks=[early_stop])
```

```python
loss,acc = model.evaluate(x_test,y_test)
print("Test Accuracy", acc)
loss
```
