```python
import os
import numpy as np
from scipy.io import wavfile
from python_speech_features import mfcc
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib
```

```python
dataset_path = "mini_speech_commands"
```

```python
def extract_features(file_path):
    rate,signal = wavfile.read(file_path)
    print(signal)

    if signal.ndim > 1:
        signal = signal[:, 0]

    features = mfcc(signal, rate, numcep=40)
    return np.mean(features, axis=0)
```

```python
x,y = [],[]

for label in os.listdir(dataset_path):
    label_path =  os.path.join(dataset_path,label)

    if os.path.isdir(label_path):
        for file in os.listdir(label_path):
            if file.endswith(".wav"):
                file_path = os.path.join(label_path, file)
                x.append(extract_features(file_path))
                y.append(label)

x = np.array(x)
y = np.array(y)
print("Samples loaded:", len(x))
```

```python
encoder = LabelEncoder()
y_encoding = encoder.fit_transform(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y_encoding, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier(n_estimators=200)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
print("Accuracy:",accuracy_score(y_test,y_pred))
```
